```python
import numpy as np
from sklearn.datasets import make_circles
```

```python
from sklearn.model_selection import train_test_split
```

```python
X,y=make_circles(n_samples=1000,noise=0.4,factor=0.1,random_state=0)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X_train,y_train,dt)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
```

```python
rf.fit(X_train,y_train)
```

```python
plot_decision_regions(X_train,y_train,rf)
```

```python
def f(x):
    x=x.ravel()
    return np.exp(-x**2)+1.5*np.exp(-(x-2)**2)
```

```python
def gen(n_samples,noise):
    X=np.random.rand(n_samples)*10-5
    X=np.sort(X).ravel()
    y=np.exp(-X**2)+1.5*np.exp(-(X-2)**2)+np.random.normal(0.0,noise,n_samples)
    X=X.reshape((n_samples,1))
    return X,y
```

```python
X_train,y_train=gen(100,0.1)
```

```python
X_test,y_test=gen(150,0.1)
```

```python
plt.plot(X_train,f(X_train),c='red')
plt.scatter(X_train,y_train,c='blue',s=5)
```

```python
plt.plot(X_test,f(X_test))
plt.scatter(X_test,y_test,c='red',s=5)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
dr=DecisionTreeRegressor()
rr=RandomForestRegressor()
```

```python
dr.fit(X_train,y_train)
rr.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
y_pred_dr=dr.predict(X_test)
y_pred_rr=rr.predict(X_test)
```

```python
r2_score(y_test,y_pred_dr)
```

```python
r2_score(y_test,y_pred_rr)
```

```python
mean_squared_error(y_test,y_pred_dr)
```

```python
mean_squared_error(y_test,y_pred_rr)
```

```python
plt.plot(X_test,f(X_test),c='r')
plt.scatter(X_train,y_train,c='b',s=5)
plt.plot(X_test,y_pred_dr,linewidth=1)
```

```python
plt.plot(X_test,f(X_test),c='r')
plt.scatter(X_train,y_train,c='b',s=5)
plt.plot(X_test,y_pred_rr,linewidth=1)
```
