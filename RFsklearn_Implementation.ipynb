```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\ASUS\Downloads\datasets\heart.csv")
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV,RandomizedSearchCV
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
df
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.drop(columns='target'),df['target'],random_state=2,test_size=0.2)
```

```python
param_grid={'n_estimators':[1,10,50,100,150,500],'max_features':[1,3,5,7,9,11,13],'max_samples':[0.25,0.5,0.75,1],'max_depth':[2,5,8,None]}
```

```python
rf=RandomForestClassifier()
```

```python
g=GridSearchCV(rf,cv=3,param_grid=param_grid,scoring='accuracy')
```

```python
g.fit(X_train,y_train)
```

```python
g.best_params_
```

```python
g.best_score_
```

```python
g.cv_results_
```

```python
rf_gscv=RandomForestClassifier(n_estimators=100,max_depth=None,max_samples=0.75,max_features=1)
```

```python
rf_gscv.fit(X_train,y_train)
y_pred=rf_gscv.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
rf2=RandomForestClassifier()
r=RandomizedSearchCV(rf2,param_distributions=param_grid,verbose=2,n_jobs=-1,cv=3)
r.fit(X_train,y_train)
```

```python
r.best_params_
```

```python
r.best_score_
```

```python
rf_rscv=RandomForestClassifier(n_estimators=100,max_samples=0.75,max_features=9,max_depth=8)
rf_rscv.fit(X_train,y_train)
y_pred_rscv=rf_rscv.predict(X_test)
accuracy_score(y_test,y_pred_rscv)
```

```python
rf_oob=RandomForestClassifier(oob_score=True)
rf_oob.fit(df.drop(columns='target'),df['target'])
rf_oob.oob_score_
```

```python
y_pred_oob=rf_oob.predict(X_test)
accuracy_score(y_test,y_pred_oob)
```
